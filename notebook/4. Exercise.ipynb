# Angenohmen wir erhöhen 20 auf 100

```python
test_2 = pd.Series([1, 2, 3, 4, 100])
```

```python
test_2.describe()
```

```python
mean_2 = test_2.describe()['mean']

mean_2
```

```python
s = test_2.describe()['std']
s
```

```python
test_2 - mean_2
```

```python
( test_2 - mean_2 ) / s
```

```python
test_after_scaling = ( test_2 - mean_2 ) / s

test_after_scaling
```

### Achte hier auf: mean     0.000000 ist somit viel verkleinert
