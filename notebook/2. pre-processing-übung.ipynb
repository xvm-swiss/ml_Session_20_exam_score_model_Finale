# Encoding Categorical Columns

### * Encoding ordinal columns using Label Encoding

```python
df.sample()
```

### columns to label encode :
 'sleep_quality', 'facility_rating', 'exam_difficulty'

```python
df['exam_difficulty'].unique()
```

```python
def encode_sleep_quality( df ) :
    df['sleep_quality'] = df['sleep_quality'].map({
        'poor' : 0, 'average' : 1, 'good' : 2
    })
    return df
    
df = encode_sleep_quality( df )    
```

```python
def encode_facility_rating( df ) :
    df['facility_rating'] = df['facility_rating'].map({
        'low' : 0, 'medium' : 1, 'high' : 2
    })
    return df
    
df = encode_facility_rating( df )    
```

```python
def encode_exam_difficulty( df ) :
    df['exam_difficulty'] = df['exam_difficulty'].map({
        'easy' : 0, 'moderate' : 1, 'hard' : 2
    })
    return df
    
df = encode_exam_difficulty( df )    
```
